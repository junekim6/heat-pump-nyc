## Heat Pump usage for heating, by state

```python
# This data is from 2020 RECS Survey Data from U.S. Energy Information Administration 
# Link: https://www.eia.gov/consumption/residential/data/2020/index.php?view=state

df = pd.read_excel("State Space Heating.xlsx", na_values=0)
```

```python
df.drop(df.columns[[8, 9, 10, 11]], axis=1, inplace=True)
df.drop([0, 1, 2, 3, 4, 5, 67], axis=0, inplace=True)
```

```python
df['Unnamed: 5'] = pd.to_numeric(df['Unnamed: 5'], errors='coerce')
```

```python
df.head()
```

```python
df.rename(columns={
    'Preliminary data release date: June 2022':'state',
    'Unnamed: 1':'pop_in_mill',
    'Unnamed: 2':'furnace_mill',
    'Unnamed: 3':'furnace_pct',
    'Unnamed: 4':'hp_mill',
    'Unnamed: 5':'hp_pct',
    'Unnamed: 6':'steam_hot_water_boiler_mill',
    'Unnamed: 7':'steam_hot_water_boiler_pct'
    }, inplace=True)
```

```python
df.state = df.state.str.strip()
```

```python
df.state.unique()
```

```python
df_states = df[df.state.notnull()]
df_states.hp_pct = df_states.hp_pct.fillna(0)
df_states
```

```python
states_hp = df_states[df_states.columns[[0, 5]]]
```

```python
states_hp.to_csv('states_heat_pump.csv', index=False)
```

## New York City GHG inventory

```python
nyc_ghg_df = pd.read_csv("NYC_GHG_Inventory.csv")
```

```python
nyc_ghg_df.head(10)
```

```python
nyc_ghg_df.drop(nyc_ghg_df.columns[[6, 7, 8, 9]], axis=1, inplace=True)
```

```python
# Rename columns

nyc_ghg_df.columns = ['sector', 'category', 'source', 'unit', 'consumed', 'tCo2e']
```

```python
# Drop rows without values

nyc_ghg_df.drop([0, 1], axis=0, inplace=True)
```

```python
nyc_ghg_df.head(10)
```

```python
nyc_ghg_df.tCo2e = nyc_ghg_df.tCo2e.astype(float)
```

```python
co2_by_category_df = pd.DataFrame(nyc_ghg_df.groupby('category')['tCo2e'].sum().reset_index())
```

```python
co2_by_category_df.info()
```

```python
co2_by_category_df.to_csv('category_co2.csv', index=False)
```
